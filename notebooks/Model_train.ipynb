```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
## Independent and depemdent features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
df = df.dropna(subset=['FWI']).reset_index(drop=True)
```

```python
y = df['FWI']
y.isna().sum()
```

```python
X
```

```python
X = df.drop(['FWI', 'Classes'], axis=1)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(15,15))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
high_corr_features = correlation(X_train, 0.85)
high_corr_features
```

```python
## drop features when correlation is more than 0.85
X_train.drop(high_corr_features,axis=1,inplace=True)
X_test.drop(high_corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

### Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python
X_train_scaled
```

## Box plot to understand the effect of Standard Scalar

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
y.isna().sum()
y[y.isna()]
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linereg = LinearRegression()
linereg.fit(X_train_scaled,y_train)
y_pred=linereg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute Error :",mae)
print("R2 Score :",score)
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute Error :",mae)
print("R2 Score :",score)
plt.scatter(y_test,y_pred)
```

## Lasso CV

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute Error :",mae)
print("R2 Score :",score)
plt.scatter(y_test,y_pred)
```

## Ridge CV

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute Error :",mae)
print("R2 Score :",score)
plt.scatter(y_test,y_pred)
```
